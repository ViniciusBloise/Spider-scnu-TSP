```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(101,500, 400)
y = np.ceil(500000 / x)
plt.title('# instances per problem size')
plt.xlabel('problem size', color='black')
plt.ylabel('# instance', color='black')
plt.plot(x, y, '-r',label='# instances per problem size')
plt.grid()


```
